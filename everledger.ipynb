```python

df = pd.read_excel('/kaggle/input/employee/employee__1_.xls',converters={'Date of Joining':str,'Date of Birth':str})
```

```python
from datetime import datetime

x =[]
for d in df["Date of Birth"]:
    if "/" in d:
        x.append(datetime.strptime(d, "%m/%d/%Y").strftime("%Y-%m-%d"))
    else:
        x.append(datetime.strptime(d,"%Y-%m-%d %H:%M:%S" ).strftime("%Y-%m-%d"))
```

```python
y=[]
for d in df["Date of Joining"]:
    if "/" in d:
        y.append(datetime.strptime(d, "%m/%d/%Y").strftime("%Y-%m-%d"))
    else:
        y.append(datetime.strptime(d,"%Y-%m-%d %H:%M:%S" ).strftime("%Y-%m-%d"))
```

```python
df.drop(columns =['Date of Joining', 'Date of Birth'])
```

```python
df["Date of Joining"]=y
```

```python
df["Date of Birth"]=x
```

```python
gb = df.groupby("Quarter of Joining")
```

```python
for key, value in gb:
    print(key)
    print(value.sort_values(by="Date of Birth", ascending=True))
    
```
